#Silver Notebook lookup tables

###Paramters

```python
dbutils.widgets.text('sourcefolder','netflix_directors')
dbutils.widgets.text('targetfolder','netflix_directors')
```

###Variables

```python
var_src_folder  = dbutils.widgets.get('sourcefolder')
var_tgt_folder  = dbutils.widgets.get('targetfolder')
```

```python
df = spark.read.format('csv')\
       .option('header',True)\
       .option('inferSchema',True)\
       .load(f'abfss://bronze@netflixprojectdatalake.dfs.core.windows.net/{var_src_folder}')
```

```python
df.limit(10).display()
```

```python
df.write.format('delta')\
      .mode('append')\
      .option('path',f'abfss://silver@netflixprojectdatalake.dfs.core.windows.net/{var_tgt_folder}')\
      .save()
```
